```python
@app.get("/patient/{patient_id}")  #Dynamic url
def view_patients(patient_id):
    data = load_data()
    return data[patient_id]


### Path function
Path() function is used to add metadata, validation rules, and documentation hints for path parameters in your api endpoints 

example:   
Title,  
description,  
min_length, max_length,  
ge(greater than equal to), gt(greater than), le(less than equal to), lt(less than)    
regex


```python 
from fastapi import FastAPI, Path
def view_patients(patient_id: str = Path(..., description="Enter the patient_id", example="P001")):
    #your function here

### HTTP status codes

Here is what happens:  
client <----> your website in a server,AND this two way communication follows HTTP protocol   




HTTP status codes indicate the result of an HTTP request
| Range   | Class        | Meaning                                                             |
| ------- | ------------ | ------------------------------------------------------------------- |
| **2xx** | Success      | Request was received, understood, and processed successfully        |
| **3xx** | Redirection  | Client must take additional action to complete request              |
| **4xx** | Client Error | Problem caused by the client (bad request, unauthorized, not found) |
| **5xx** | Server Error | Problem occurred on the server side                                 |


#### _Common HTTP Status Codes (API-focused)_

| Status Code | Class | Meaning | Typical API Use Case |
|------------|------|---------|----------------------|
| 200 | OK | Standard success | Successful GET or POST |
| 201 | Created | Resource created successfully | After POST that creates data |
| 202 | Accepted | Request accepted, processing later | Background task started |
| 204 | No Content | Success, no response body | After DELETE request |
| 301 | Moved Permanently | Resource moved permanently | URL change (mostly browsers) |
| 302 | Found | Temporary redirect | Temporary URL redirection |
| 304 | Not Modified | Cached version is valid | Client-side caching |
| 400 | Bad Request | Malformed or invalid request | Missing field, wrong data type |
| 401 | Unauthorized | Authentication required or invalid | Login required / invalid token |
| 403 | Forbidden | Authenticated but not allowed | Logged in but no permission |
| 404 | Not Found | Resource does not exist | Patient ID not found |
| 405 | Method Not Allowed | HTTP method not supported | POST on a GET-only endpoint |
| 409 | Conflict | Resource conflict | Duplicate patient ID |
| 422 | Unprocessable Entity | Validation error | FastAPI input validation failure |
| 500 | Internal Server Error | Generic server failure | Bug or crash on server |
| 502 | Bad Gateway | Invalid response from backend | Nginx can't reach API server |
| 503 | Service Unavailable | Server overloaded or down | Server maintenance |
| 504 | Gateway Timeout | Backend took too long | Slow database / API timeout |


`HTTPException` is a special built-in FastAPI exception that is used to return custom HTTP error responses when something goes wrong in the API.   

So instead of just throwing the error we can gracefully __raise an error__ with:
- Proper HTTP status code
- Custom message
- Extra headers

### Query parameters 
Query parameters are optional key-value pairs that are appended in URL to pass additional data to the server in an HTTP request. There are typically used for filtering, sorting, searching, pagination without altering the url itself

ex: 
`/patients?city=Delhi&sort_by=age`

- query params are started from "?" 
- different key-value pairs are separated by "&"

`Query()` is a utility function provided by **FastAPI** to **declare, validate, and document query parameters** in your API endpoints.

It allows you to:
- Set **default values**
- Enforce **validation rules**
- Add **metadata** like description, title, examples


Common `Query()` Parameters

| Parameter | Purpose |
|---------|---------|
| `default` | Set default value (e.g., `Query(0)`) |
| `title` | Displayed in API docs |
| `description` | Detailed explanation in Swagger |
| `example` / `examples` | Provide sample inputs |
| `min_length`, `max_length` | Validate string length |
| `ge`, `gt`, `le`, `lt` | Validate numeric bounds |
| `regex` | Pattern match for strings |
