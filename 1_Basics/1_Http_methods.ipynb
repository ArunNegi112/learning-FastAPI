There are two types of Website: `Static` and `Dynamic`
- Static are those with which user do not interact   
ex: blogs, articles
- Dynamic are those with which user interacts
ex: instagram, any tools website (majority of the websites basically)

And there are 4 types of interations possible with any website: ___CRUD___


For these interations we have different HTTP methods
- Create --> POST
- Retrieve --> GET
- Update --> PUT
- Delete --> DELETE
