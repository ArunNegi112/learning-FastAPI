### What is API?
APIs are like connectors, that enable frontend and backend of an application to communicate with each other using defined set of rules, protocols and data format

### Need for API
Earlier the applications created were based on 'Monolithic Architecture'   
`Monolithic architecture`: means the whole app is built as a single unit. Database, backend, frontend all together in a single folder tightly coupled and deployed together. A single small change in one file can cause the whole app to be reviewed and rebuild. 
This is okay for small applications but the problems are:
- Tight coupling
- Scalability 
- Slow development
- Deployment risk
- Bad collaboration scope

API solves it as,  
We can create backend and frontend separately and use API to establish communication between them.  
Also and now, not just one frontend but anyone can use the API to communicate with backend in the same manner. 

The only thing changes for machine learning applications is that, we have a ML model instead of database

### FastAPI 
A high performace web-framework for building API using python

It is based on two existing libraries in python:
- Starlette: manages how api sends request and recieve responses
- Pydantic: used to check if the data coming to api is correct and in the right format

Philosophy of FastAPI 
- Fast to run
- Fast to code 

1. Why fast to run?
How a request goes to API:   
`Web server --> SGI (Server Gateway Interface) --> API code`   

Web server is a program that listens to http request, forwards them, sends back http response  
SGI is like a translator between API(python code) and web server(can be any other language), it converts the request in the format that api can understand.

Comparing Flask and FastAPI

The thing is,   
Gunicorn typically runs applications using the WSGI interface, which is synchronous.
Uvicorn runs applications using the ASGI interface, which supports asynchronous execution.


synchronymous: Can handle only one task at a time, if one task takes time, all task waits    
example-  
Task A starts  
Task A finishes  
Task B starts  
Task B finishes  

example function- 
```python
def get_data():
    data = slow_database_call()  # blocks everything
    return data
```

Asynchronymous: Can handle multiple tasks at a time   
exmaple-  
Task A starts  
Task A waits (I/O)  
Task B runs  
Task A resumes  

example function- 
```python
async def get_data():
    data = await slow_database_call()
    return data
```

2. Why fast to code?  
- Built using Pydantic, which automatically checks input format and handle input validation
- Seamless integration with modern ecosystem (kubernetes, sqlalchemy, docker, ML/DL libraries etc)
- Auto-generated interative documentation- It automatically generates an interative document of the software we creating (so our time wont be wasted on documentation)

### Basic concepts

What is web server?  
A web server is a system that store, processes, and delivers website HTTP/HTTPS request and responses to the user over the internet

What is Gunicorn?  
Gunicorn is a Python WSGI HTTP server. It accepts HTTP request, creates worker processes, send request to Flask, and finally return responses

What is Uvicorn?  
Just like Gunicorn runs Flask, Uvicorn runs FastAPI

_So hear me out, Flask or FastAPI apps are just python applications - gunicorn/uvicorn are the wsgi/asgi server that accepts the http request and forward them to the application and return the responses_

What is WSGI?   
Web Server Gateway Interface  
A standard which defines - how a web server sends request to python application (or API), how application should return a response, and what format the response should have. etc etc... In short, it is a standard (set of rules) that is applied to a web server (Gunicorn)

What is ASGI?  
Asynchronous Server Gateway Interface    
A replacement for WSGI.  
It supports - HTTP, Web Sockets, Async programming, handle multiple requests simultaneously 


| WSGI                           | ASGI                                |
| ------------------------------ | ----------------------------------- |
| Synchronous                    | Asynchronous                        |
| Handles one request per worker | Can handle many concurrent requests |
| Flask                          | FastAPI                             |
| Gunicorn                       | Uvicorn                             |
| No WebSocket support           | WebSocket support                   |
| Older standard                 | Modern standard                     |


Why Flask uses WSGI but FastAPI uses ASGI?    
Because Flask was created in 2010, when Python only had synchronous programming.  
Python now supports async, await - so FastAPI was built on ASGI.

### Basic Computer Network concepts 
1. Throughput: The volume of data processed in a given amount of time 
2. Latency: The delay it takes for a single unit of data to reach destination from source
3. Jitter: The variation in latency over time is called jitter. For a smooth application we need to have minimum jitter
4. Concurrency: The number of requests/tasks being processed simultaneously 


_Async programming does not magically improve the speed of application they only improves the throughput and concurrency of the application_
